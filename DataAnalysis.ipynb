# 收益率折线图

## 数据拼接

```python
gzmt['trade_date']=pd.to_datetime(gzmt['trade_date'])
gzmt = gzmt.set_index('trade_date')
```

```python
ndsd['trade_date']=pd.to_datetime(ndsd['trade_date'])
ndsd = ndsd.set_index('trade_date')
```

```python
stock2 = pd.concat([gzmt['close'],ndsd['close']],axis=1)
stock2.columns=['贵州茅台','宁德时代']
```

```python
stock2.plot()
```

## 缺失值查看

```python
stock2.info()
```

```python
stock2.isnull().sum()
```

### 收益率计算

```python
stock2=stock2.sort_index(ascending=True)
```

```python
stock2.head(5)
```

```python
stock2.shift(1)
```

```python
R=np.log(stock2/stock2.shift(1))
```

### 缺失值去除

```python
R = R.dropna()
```

```python
R
```

## 收益率画图

### 收益率统计描述

```python
R.describe()
```

### 对数收益率

```python
R.plot()
```

### 直接计算收益率

```python
(np.exp(R)-1).plot()
```

### 查看差距

```python
((np.exp(R)-1)-R).describe()
```

### 修改蜡烛图

https://blog.csdn.net/Shepherdppz/article/details/117575286

```python
import mplfinance as mpf
# 设置mplfinance的蜡烛颜色，up为阳线颜色，down为阴线颜色
my_color = mpf.make_marketcolors(up='r',
                                 down='g',
                                 edge='inherit',
                                 wick='inherit',
                                 volume='inherit')
# 设置图表的背景色
my_style = mpf.make_mpf_style(marketcolors=my_color,
                              figcolor='(0.82, 0.83, 0.85)',
                              gridcolor='(0.82, 0.83, 0.85)')
```

```python
mpf.plot(gzmt,type='candle',style=my_style,volume=True)
```
